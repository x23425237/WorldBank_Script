## Repayment Rate  All Regions in the Dataset 
* LATIN AMERICA AND CARIBBEAN- Highest repayment rate of 95% 
* EUROPE AND CENTRAL ASIA - lowest of 70% 



```python
 # https://www.datacamp.com/tutorial/tutorial-postgresql-python
    
    # Define the SQL query to execute
# Define the SQL query
query = """
SELECT region, 
       SUM(disbursed_amount_usd) AS total_disbursed,
       SUM(repaid_to_ibrd_usd) AS total_repaid,
       SUM(repaid_to_ibrd_usd) / SUM(disbursed_amount_usd) * 100 AS repayment_rate
FROM loan_data
WHERE loan_status = 'Fully Repaid'
GROUP BY region
ORDER BY repayment_rate DESC;
"""

# Execute the query
cursor.execute(query)

# Fetch the results
result = cursor.fetchall()

# Prepare the data for plotting
regions = [row[0] for row in result]
repayment_rates = [row[3] for row in result]

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x=regions, y=repayment_rates, palette="viridis")

# Customize the plot
plt.title('Repayment Rate by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Repayment Rate (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate region names for better readability
plt.tight_layout()  # Adjust layout to prevent overlap

# Display the plot
plt.show()
```

```python
# https://medium.com/@alestamm/importing-data-from-a-postgresql-database-to-a-pandas-dataframe-5f4bffcd8bb2

# sum of disbursed_amount_usd for each country in EASTERN AND SOUTHERN AFRICA

# Define the SQL query
query = """
select cast(sum(disbursed_amount_usd) as bigint) as total_disbursed,
       cast(sum(repaid_to_ibrd_usd) as bigint) as total_repaid,
       country_economy 
from loan_data
where region = 'EASTERN AND SOUTHERN AFRICA'
group by country_economy
order by total_disbursed desc
limit 10;
"""

# Execute the query
cursor.execute(query)

# Fetch the results
results = cursor.fetchall()

# Convert the results to a DataFrame
df = pd.DataFrame(results, columns=['total_disbursed', 'total_repaid','country_economy'])

# Display the DataFrame
print(df)
```

```python
# https://plotly.com/python/line-charts/

# SQL query
query = """
WITH cte AS (
    SELECT 
        w.country_name,
        l.disbursed_amount_usd,
        w.unemployment_rate,
		w.region
    FROM world_eco_indicator w
    LEFT JOIN loan_data l 
        ON w.country_name = l.country_economy
)
SELECT 
    cast(AVG(cte.disbursed_amount_usd) as int) AS avg_disbursed_amount_usd,
    round(AVG(cte.unemployment_rate),2) AS avg_unemployment_rate,
    cte.country_name,
	cte.region
FROM cte
where cte.disbursed_amount_usd is not null
and cte.region='Sub-Saharan Africa'
GROUP BY cte.country_name,cte.region
"""
# Execute the query
cursor.execute(query)

# Fetch the results
results = cursor.fetchall()

# Convert the results to a DataFrame
df = pd.DataFrame(results, columns=['avg_disbursed_amount_usd', 'avg_unemployment_rate','country_name','region'])
df.head()

```

```python
# https://medium.com/@msjahid/line-plot-or-line-chart-in-python-with-math-ce154a7e9b17


# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the bar graph for 'avg_disbursed_amount_usd'
ax1.bar(df['country_name'], df['avg_disbursed_amount_usd']/ 1e9, color='b', alpha=0.6, label='Disbursed Amount (USD)')

# Create a second y-axis to plot 'avg_unemployment_rate'
ax2 = ax1.twinx()
ax2.plot(df['country_name'], df['avg_unemployment_rate'], color='r', marker='o', label='Unemployment Rate (%)')

# Add titles and labels
ax1.set_title('Disbursed Amount and Unemployment Rate by Country')
ax1.set_xlabel('Country')
ax1.set_ylabel('Disbursed Amount in billion (USD)', color='b')
ax2.set_ylabel('Unemployment Rate (%)', color='r')

# Rotate x-axis labels for better visibility
ax1.tick_params(axis='x', rotation=45)  # This rotates the x-axis labels

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# pie chart 

# SQL query
query = """
SELECT 
        w.incomegroup,
	cast(AVG(l.disbursed_amount_usd) as int) AS avg_disbursed_amount_usd
    FROM world_eco_indicator w
    LEFT JOIN loan_data l 
        ON w.country_name = l.country_economy
group by w.incomegroup
order by 2 desc
"""
# Execute the query
cursor.execute(query)

# Fetch the results
results = cursor.fetchall()

# Convert the results to a DataFrame
df = pd.DataFrame(results, columns=['Income_group', 'avg_disbursed_amount_usd'])
df.head()
```

```python
# Calculate total disbursed amount
total_disbursed = df['avg_disbursed_amount_usd'].sum()

# Calculate the percentage of the total for each income group
df['percentage'] = round((df['avg_disbursed_amount_usd'] / total_disbursed) * 100,2)


# Create a pie chart
fig = go.Figure(data=[go.Pie(
    labels=df['Income_group'],  # The income groups
    values=df['percentage'],  # The average disbursed amounts
    hoverinfo='label+percent',  # Show both label and percentage on hover
    textinfo='label+value',  # Show both label and value on the chart
    texttemplate='%{value:.2f}%',
)])

# Add title
fig.update_layout(
    title="Average Percentage Disbursed Amount by Income Group (USD)"
)

# Show the figure
fig.show()
```
